File descriptor greater than 2 are assigned to files we open or work with.

```python
#include <unistd.h>
#include <fcntl.h>
#include <sys/types.h>
#include <sys/stat.h>

int main() {
    int n, fd;

    char write_data[] = "Samiul Kadir";
    fd = open("write_file.txt", O_WRONLY); // write_file.txt should exist.
    // if write_file.txt doesn't exist then create it first using touch.


    // what if you dont want to manually create the file. Then we use O_CREAT.
    // fd = open("write_file.txt", O_WRONLY | O_CREAT,0644);
    write(fd, write_data , sizeof(write_data));
	  close(fd);

    return 0;
}
```

## read()

 In this code we are getting the output of what is written in `write_file.txt.`


```python
#include<unistd.h>
#include<fcntl.h>
#include<sys/types.h>
#include<sys/stat.h>

int main () {
	int n, fd;
	char read_data[50];
	fd= open("write_file.txt", O_RDONLY);
	n = read(fd, buffer, 10);
	printf("The read data is %s",read_data);
}
```

We can use O_CREAT mode to create the files automatically if the file does not exist. In this code we are writing in `fd1` which refers to the `input.txt` file.

`fd2 = open("output.txt", O_CREAT | O_WRONLY, 0642);`

The `0642` is the permission for user, group and others needed for creating the file with `O_CREAT`


```python
#include<unistd.h>
#include<fcntl.h>
#include<sys/types.h>
#include<sys/stat.h>

int main () {
	int n, fd1, fd2;
	char buffer[50];
	fd1= open("input.txt", O_RDONLY);
	n = read(fd1, buffer, 50);
	fd2 = open("output.txt", O_CREAT | O_WRONLY, 0642);
	write(fd1, buffer, n);
}
```

## lseekâ†’ When we want to move the pointer/cursor

`read()` always starts from the 1st character in the file.

`write()` happens from the beginning

seeking.txt â†’ `abcdHello1234World`

```python
#include<unistd.h>
#include<fcntl.h>
#include<sys/types.h>
#include<sys/stat.h>

int main () {
	int n, fd1, fd2;
	char buffer[18];
	fd1 = open("seeking.txt", O_RDWR); //abcdHello1234World
	read(fd1 , buffer, 4); //abcd
	write(1, buffer, 4);
	read(fd1, buffer, 5); //Hello, reads the next 5 character as indicated by bytes read
	write(1, buff, 5);

}
```

```python
#include<unistd.h>
#include<fcntl.h>
#include<sys/types.h>
#include<sys/stat.h>

int main () {
	int n, fd1, fd2;
	char buffer[18];
	fd1 = open("seeking.txt", O_RDWR); //abcdHello1234World
	read(fd1 , buffer, 4); //abcd, pointer is in H
	write(1, buffer, 4);
	lseek(fd1, 9, SEEK_CUR); //Pointer moves 9 chars ahead of H and is in W now
	read(fd1, buffer, 5); //World, reads the next 5 character as indicated by bytes read
	write(1, buffer, 5);

}
```

When we read(fd1 , buffer, 4); read 4 bytes the pointer is in H, SEEK_CUR will move lseek(fd1, 9, SEEK_CUR); 9 characters ahead, and print the remaining words.

```python
#include<unistd.h>
#include<fcntl.h>
#include<sys/types.h>
#include<sys/stat.h>

int main () {
	int n, fd1, fd2;
	char buffer[18];
	fd1 = open("seeking.txt", O_RDWR); //abcdHello1234World
	fd2 = lseek(fd1, 9, SEEK_SET);
	printf("Pointer is at %d position\n", fd2);//Pointer is at 10 position
	read(fd1 , buffer, 9); //1234World, reads the next 9 character as indicated by bytes read
	write(1, buffer, 9);
}
```

SEEK_SET starts from beginning of the file.

```python
#include<unistd.h>
#include<fcntl.h>
#include<sys/types.h>
#include<sys/stat.h>

int main () {
	int n, fd1, fd2;
	char buffer[18];
	fd1 = open("seeking.txt", O_RDWR); //abcdHello1234World
	fd2 = lseek(fd1, -6, SEEK_END);// moves the poineter from the end of the file, also include 0/ end of string character.
	read(fd1 , buffer, 5); //World
	write(1, buffer, 5);
}
```

## Fork and exec system calls

- It's only available in linux OS.

<aside>
ðŸ’¡The fork() system call is used to create a seperate, duplicate process

</aside>

The process ID is different, parent and child process.

<aside>
ðŸ’¡ When an exec() system is invoked, the program specified in the parameter to exec() will replace the entire process.

</aside>

Have same PID, replace an existing one not creating a new process

```python
#FORK
#include <stdio.h>
#include <stdlib.h>
#include <sys/wait.h>
#include <unistd.h>

int main() {
    pid_t pid;
    int status;

    pid = fork(); // Create a new process

    if (pid == -1) {
        // If fork() returns -1, an error occurred
        perror("fork failed");
        exit(EXIT_FAILURE);
    } else if (pid == 0) {
        // This is executed by the child process
        printf("Child process: PID = %d\n", getpid());
        printf("Performing some task...\n");
        sleep(2); // Simulate a task
        printf("Child process is done.\n");
        exit(EXIT_SUCCESS); // Child process exits
    } else {
        // This is executed by the parent process
        printf("Parent process: PID = %d\n", getpid());
        printf("Waiting for child process to finish...\n");
        wait(&status); // Wait for the child process to finish
        printf("Parent process is done.\n");
    }

    return EXIT_SUCCESS;
}
```
